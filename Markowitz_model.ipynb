```python
import pandas_datareader.data as web
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# start date and end date
start = datetime.datetime(2021, 1, 1)
end = datetime.datetime(2021, 1, 22)
```

```python
# get historical stock data using 'stooq' api ('yahoo' doesn't work)
ceva = web.DataReader('CEVA', 'stooq', start, end)
google = web.DataReader('GOOGL', 'stooq', start, end)
tesla = web.DataReader('TSLA', 'stooq', start, end)
zom = web.DataReader('ZOM', 'stooq', start, end)
```

```python
# get closing prices of stocks of those companies
stocks = pd.concat([ceva['Close'], google['Close'], tesla['Close'], zom['Close']],axis=1)
```

```python
# rename columns
stocks.columns = ['CEVA', 'GOOGLE', 'TESLA', 'ZOMEDCA']
stocks
```

### Returns:
$$ r_t = \frac{X_t}{X_{t-1}} $$
##### where $X_t$ is the stock price at time $t$

```python
returns = stocks/stocks.shift(-1)
returns
```

### Log Returns
##### log returns introduces the fairness
$$ \log(r_t) = \log(\frac{X_t}{X_{t-1}}) $$

```python
logReturns = np.log(returns)
pBar = logReturns.mean()
Sigma = logReturns.cov()
logReturns
```

### Sharpe Ratio
$$ SR(w) = \frac{R(w) - R_f}{\sigma(w)} $$
##### where $w$ is the weight matrix, and $R_f$ is the risk free rate
$$ R(w) = w^T\log(r) $$
$$ \sigma(w) = \sqrt{w^T\Sigma w} $$
##### where the $\Sigma$ is the covariance
##### so the Sharpe Ratio is proportional to:
$$ SR(w) \sim \frac{w^T\log(r)}{\sqrt{w^T\Sigma w}} $$
##### because the risk free rate is common for all weights

### Return vs Volatility

```python
no0fPortfolios = 10000
weight = np.zeros((no0fPortfolios ,4))
expectedReturn = np.zeros(no0fPortfolios)
expectedVolatility = np.zeros(no0fPortfolios)
sharpeRatio = np.zeros(no0fPortfolios)
meanLogRet = logReturns.mean()
Sigma = logReturns.cov()
for k in range(no0fPortfolios):
    
    # generate random weight vector
    w = np.array(np.random.random(4))
    w = w / np.sum(w)
    weight [k,:] = w
    
    # Expected log return
    expectedReturn[k] = np.sum(meanLogRet * w )
    
    # Expected volatility
    expectedVolatility[k] = np.sqrt(np.dot(w.T, np.dot(Sigma,w)))
    
    # Sharpe Ratio
    sharpeRatio[k] = expectedReturn[k]/expectedVolatility[k]
```

```python
# get the weights that maximises the Sharpe Ratio
index = sharpeRatio.argmax()
weight[index, :]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,6))
plt.scatter (expectedVolatility, expectedReturn, c=sharpeRatio)
plt.xlabel('Expected Volatility')
plt.ylabel('Expected Log Returns')
plt.colorbar (label= 'SR')
# the optimal point is labelled red
plt.scatter(expectedVolatility[index], expectedReturn[index] , c='red')
plt.show()
```

### Efficient Frontier
##### optimisation problem:
$$
\begin{aligned}
\text{minimise} \quad &w^T\Sigma w \\
\text{s.t.} \quad &\bar{r}^T w = \bar{r_p} \\
&w \geq 0 \\
&1^T w = 1
\end{aligned}
$$
##### or use the Sharpe Ratio
$$
\begin{aligned}
\text{minimise} \quad &-SR(w) \\
\text{s.t.} \quad &0\leq w_i \leq1
\end{aligned}
$$

```python
from scipy.optimize import minimize

def negativeSR(w):
    w = np.array(w)
    R = np.sum(meanLogRet*w)
    V = np.sqrt(np.dot(w.T, np.dot(Sigma,w)))
    SR = R/V
    return -1*SR

def checkSumToOne(w):
    return np.sum(w) - 1

w0 = [0.25, 0.25, 0.25, 0.25]
bounds = ((0,1), (0,1) , (0,1), (0,1))
constraints = ({'type':'eq', 'fun':checkSumToOne})

# compute the optimal solution
# (this is the second way of computing the optimal solution, the first solution is using large amount of samples)
w_opt = minimize(negativeSR, w0, method='SLSQP', bounds=bounds, constraints=constraints)
w_opt.x
```

```python
returns = np.linspace (0,0.07,50)
volatility_opt = []

def minimizeMyVolatility(w):
    w = np. array (w)
    V = np.sqrt (np.dot(w.T, np.dot(Sigma,w)))
    return V

def getReturn(w):
    w = np.array(w)
    R = np.sum(meanLogRet*w)
    return R

for R in returns:
    # find best volatility
    constraints = ({'type': 'eq', 'fun': checkSumToOne},
                   {'type': 'eq', 'fun': lambda w: getReturn (w) - R})
    opt = minimize(minimizeMyVolatility, w0, method='SLSQP' ,bounds=bounds, constraints=constraints)
    # save my optimal volatility
    volatility_opt.append(opt['fun'])
    
plt.figure(figsize=(20,6))
plt.scatter (expectedVolatility, expectedReturn, c=sharpeRatio)
plt.xlabel('Expected Volatility')
plt.ylabel('Expected Log Returns')
plt.colorbar (label= 'SR')
plt.plot(volatility_opt, returns, '--')
plt.show()
```
